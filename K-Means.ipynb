```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
## Aproveitei as bibliotecas da antiga tarefa
```

```python
data = {
    'comprimento':  [
                     8.5, 9, 7.8, 8.2, 8.9, 9.2, 8.4, 9, 8.7, 9.1, 8.6, 8.9, 9.2, 8.8, 8.4, 9.1, 8.7, 9, 8.6, 9.2,
                     10.5, 10, 8.3, 8.2, 8.8, 10.1, 8.7, 10.9, 8.2, 10.2, 8.5, 8, 8.5, 10.3, 8.6, 10.4, 8.7, 8.5, 8.9, 10,
                     8, 7.5, 8.3, 7.8, 8.2, 7.7, 8.5, 8.1, 7.9, 8.3, 7.6, 8, 7.4, 8.1, 7.9, 8.4, 7.8, 8.2, 7.5, 8
                     ],

    'altura': [
                     8.4, 9.1, 7.9, 8.3, 8.8, 9.1, 8.3, 9, 8.6, 9.2, 8.5, 8.7, 9.2, 8.7, 8.5, 9, 8.8, 9, 8.7, 9.1,
                     4.5, 5, 4.8, 4.7, 4.3, 5.1, 4.9, 5, 4.4, 5.2, 4.5, 4.6, 4.2, 5.1, 4.7, 5.3, 4.6, 4.8, 4.4, 5,
                     12,5, 13.8, 13.5, 10.3, 11.7, 12.6, 12.2, 13.9, 10.5, 11.3, 10.1, 11, 14, 11.5, 14.5, 11.3, 10.9, 11.4, 10, 11.1
               ]
}
```

```python
# Transformando os dados em um array numpy
X = np.array(list(zip(data['comprimento'], data['altura'])))
```

```python
# Aplicando o K-means com 3 clusters
kmeans = KMeans(n_clusters=3).fit(X)
#kmeans.fit(X)
```

```python
# Obtendo os centroids e os labels dos clusters
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Plotando os clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', label='Dados')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='black', label='Centroids')

# Definindo o título e rótulos
plt.title('K-means Clustering - Maçã, Abacaxi e Banana')
plt.xlabel('Comprimento')
plt.ylabel('Altura')
plt.grid(True)
plt.legend()
#plt.show()

# Contagem de frutas
contagem_macas = 0
contagem_abacaxis = 0
contagem_bananas = 0

# Exibindo os clusters com as frutas e fazendo a contagem
for i, (comprimento, altura) in enumerate(X):
    if altura > comprimento:
        fruta = "Abacaxi"
        contagem_abacaxis += 1
    elif abs(comprimento - altura) <= 0.5:  # Tolerância de 0.5 para a igualdade
        fruta = "Maçã"
        contagem_macas += 1
    else:
        fruta = "Banana"
        contagem_bananas += 1

    print(f'Comprimento: {comprimento}, Altura: {altura} -> {fruta}')

# Exibindo a contagem
print("\nContagem de frutas:")
print(f"Maçãs: {contagem_macas}")
print(f"Abacaxis: {contagem_abacaxis}")
print(f"Bananas: {contagem_bananas}")

plt.show()
```
